```python
# Univariate Analysis- In Univariate analysis we analize each and every single feature.
print(df['Job_Title'].value_counts())
```

```python
df['Job_Title'].value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df,x='Industry')
plt.title('Industry')
```

```python
# Bivariate Analysis- Here we analize more than one features
df.groupby(df['Job_Title'])['Salary_USD'].sum().plot(kind='bar')
```

### Here we analyze that Data Scientist is the highest paying job in market.

```python
df.groupby(df['Industry'])['Salary_USD'].sum().plot(kind='bar')
```

### Here we conclude that Education industry pay the highest.

```python
plt.figure(figsize=(6,6))
df.groupby(df['Company_Size'])['Salary_USD'].sum().plot(kind='bar')
plt.title('Company Size')
```

### Here we analyze that small size companies pay the highest salary.

```python
df.groupby(df['Required_Skills'])['Salary_USD'].sum().plot(kind='bar')
plt.title('Skills Vs Salary')
```

### We conclude that market pay the highest salary for Python skills and second highest paying skill is Project Management.

```python
plt.figure(figsize=(6,6))
df.groupby(df['Location'])['Salary_USD'].sum().plot(kind='bar')
plt.title('Location Vs Salary')
```

### Here we can see that in San Francisco, Singapore, Sydney pay the highest salary for employees and where in Toronto, London pay the lowest salary.

```python
AI_Level=df[df['AI_Adoption_Level']=='High']
AI_Level.value_counts().sum()
```

```python
plt.figure(figsize=(10,6))
df.groupby('Industry')['AI_Adoption_Level'].value_counts().plot(kind='bar')
plt.title('Industry Vs AI Adoption Level')
```

### Here we can see that Technology industry has the highest medium AI adoption level is, where Healthcare industry has the lowest low AI adoption level is.

```python
plt.figure(figsize=(10,6))
df.groupby('Job_Title')['Job_Growth_Projection'].value_counts().plot(kind='bar')
plt.title('Job Title Vs Job Growth Projection')
```

```python
plt.figure(figsize=(13,5))
ax=sns.barplot(data=df,x='Industry',y='Salary_USD')
ax.bar_label(ax.containers[0])
```

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(data=df,x='AI_Adoption_Level')
ax.bar_label(ax.containers[0])
```
